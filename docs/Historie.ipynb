```python
import matplotlib.pyplot as plt
import numpy as np

# Set the seed for reproducibility
np.random.seed(0)

# Number of semesters
n_semesters = 10

# Generate random ECTS points for a specific student and for all students
student_ects = np.random.randint(0, 30, size=n_semesters)
average_ects = np.random.randint(0, 28, size=n_semesters)

# Calculate the confidence interval for the average ECTS points
conf_int = [average_ects - 2, average_ects + 2]

# Create a new figure
plt.figure()

# Plot the ECTS points for the specific student
plt.plot(range(1, n_semesters + 1), student_ects, marker='o', label='Specific Student')

# Plot the average ECTS points with the confidence interval
plt.fill_between(range(1, n_semesters + 1), conf_int[0], conf_int[1], color='gray', alpha=0.5, label='Average (Confidence Interval)')
plt.plot(range(1, n_semesters + 1), average_ects, color='black', label='Average')

# Add labels and a legend
plt.xlabel('Semester')
plt.ylabel('ECTS Points')
plt.legend()

# Show the plot
plt.show()
```

```python
def generate_students_data():
    # Generate random ECTS points for a specific student and for all students
    students_ects = np.random.normal(loc=20, scale=15, size=n_semesters)
    
    # Round the ECTS points to integers
    students_ects = np.round(students_ects).astype(int)
    
    # Ensure that all ECTS points are within the valid range (0 to 45)
    students_ects = np.clip(students_ects, 0, 45)
    
    # Generate random ECTS points for all students
    all_students_ects = np.random.normal(loc=20, scale=15, size=(n_students, n_semesters))

    # Round the ECTS points to integers
    all_students_ects = np.round(all_students_ects).astype(int)

    # Ensure that all ECTS points are within the valid range (0 to 45)
    all_students_ects = np.clip(all_students_ects, 0, 45)
    return student_ects, all_students_ects
```

```python
import matplotlib.pyplot as plt
import numpy as np
import scipy.stats as stats

# Set the seed for reproducibility
np.random.seed(0)

# Number of semesters
n_semesters = 10

# Number of students
n_students = 100

student_ects, all_student_ects = generate_students_data()

# Calculate the average ECTS points for all students
average_ects = np.mean(all_students_ects, axis=0)

# Calculate the confidence interval for the average ECTS points
conf_int = stats.t.interval(0.95, df=n_students-1, loc=average_ects, scale=stats.sem(all_students_ects))

# Create a new figure
plt.figure()

# Plot the ECTS points for the specific student
plt.plot(range(1, n_semesters + 1), student_ects, marker='o', label='Specific Student')

# Plot the average ECTS points with the confidence interval
plt.fill_between(range(1, n_semesters + 1), conf_int[0], conf_int[1], color='gray', alpha=0.5, label='Average (Confidence Interval)')
plt.plot(range(1, n_semesters + 1), average_ects, color='black', label='Average')

# Add labels and a legend
plt.xlabel('Semester')
plt.ylabel('ECTS Points')
plt.legend()

# Show the plot
plt.show()
```
